## Pertanyaan Analisis

### 1. Mengapa warna citra dapat terlihat tidak sesuai jika langsung ditampilkan setelah dibaca dengan OpenCV?

OpenCV membaca citra dalam format **BGR** (Blue-Green-Red), sedangkan Matplotlib menggunakan format **RGB** (Red-Green-Blue).
Ketika citra BGR ditampilkan dengan `plt.imshow()`, Matplotlib menginterpretasikan channel pertama sebagai Red (padahal itu Blue),
channel kedua sebagai Green (tetap Green), dan channel ketiga sebagai Blue (padahal itu Red). Akibatnya, warna biru dan merah tertukar
sehingga citra terlihat tidak natural.

### 2. Apa perbedaan BGR dan RGB?

BGR dan RGB adalah dua format urutan channel warna yang berbeda:
- **RGB**: Urutan channel Red → Green → Blue (umum digunakan di monitor, web, Matplotlib)
- **BGR**: Urutan channel Blue → Green → Red (digunakan oleh OpenCV secara default)

Keduanya menyimpan informasi warna yang sama, hanya urutan penyimpanannya yang berbeda.

### 3. Mengapa penting memahami format warna sebelum melakukan manipulasi citra?

Penting karena jika kita melakukan manipulasi (seperti mengakses channel tertentu, mengubah warna, atau menampilkan citra)
tanpa memperhatikan format warnanya, hasilnya bisa salah atau tidak sesuai yang diharapkan. Misalnya, jika kita ingin
memproses channel merah tetapi data tersimpan di channel ketiga (BGR), maka yang diproses adalah channel biru.
Memahami format warna memastikan setiap operasi berjalan pada channel yang benar.

## Pertanyaan Analisis - Resize

### 1. Mengapa dua citra harus memiliki ukuran yang sama sebelum dilakukan image blending?

Image blending adalah operasi aritmatika per-pixel. Operasi `I_blend = α*I1 + β*I2 + γ` membutuhkan
setiap pixel di posisi (x, y) pada citra pertama dijumlahkan dengan pixel di posisi (x, y) yang sama pada
citra kedua. Jika ukuran kedua citra berbeda, maka operasi penjumlahan per-pixel tidak dapat dilakukan
karena dimensinya tidak cocok. OpenCV akan menolak operasi blending jika ukuran tidak sama.

### 2. Apa risiko resize terhadap kualitas citra?

Risiko utama resize terhadap kualitas citra adalah:
- **Jika diperbesar (upscaling)**: Citra bisa menjadi buram/pixelated karena algoritma interpolasi harus
  "menebak" nilai pixel baru yang tidak ada di citra asli. Detail halus bisa hilang.
- **Jika diperkecil (downscaling)**: Informasi detail bisa hilang karena pixel-pixel dijumlahkan/dirata-rata.
  Tekstur halus dan objek kecil mungkin menjadi tidak jelas atau hilang sama sekali.

### 3. Apakah resize dapat mengubah proporsi objek dalam citra? Jelaskan.

Ya, jika rasio aspek (aspect ratio) tidak dipertahankan saat resize. Misalnya citra asli 800x600 (rasio 4:3)
di-resize menjadi 600x400 (rasio 3:2). Dalam contoh ini, proporsi objek akan berubah karena rasio aspeknya
tidak sama. Objek akan terlihat lebih lebar atau lebih sempit dari aslinya. Untuk mempertahankan proporsi,
kita harus menjaga rasio aspek yang sama atau melakukan padding/cropping.

Pada eksperimen ini, citra 1 (600x386) di-resize ke (600x400), sehingga proporsinya sedikit berubah.

## Bagian C: Image Blending

Image blending adalah teknik menggabungkan dua citra dengan bobot tertentu.
Persamaan: **I_blend = α * I1 + β * I2 + γ** dengan **α + β = 1**

Kita akan mencoba 3 kombinasi bobot:

```python
# Menentukan kombinasi bobot
weights = [
    (0.2, 0.8, "α=0.2, β=0.8 (Citra 2 lebih dominan)"),
    (0.5, 0.5, "α=0.5, β=0.5 (Seimbang)"),
    (0.8, 0.2, "α=0.8, β=0.2 (Citra 1 lebih dominan)")
]

# Melakukan blending untuk setiap kombinasi
blend_results = []
for a, b, desc in weights:
    result = cv2.addWeighted(img1_resize, a, img2_resize, b, 0)
    blend_results.append((a, b, desc, result))
    print(f"Blending: {desc}")

print("\nSemua blending berhasil!")
```

```python
# Visualisasi citra asli dan hasil blending
plt.figure(figsize=(15, 10))

# Citra 1
plt.subplot(2, 3, 1)
plt.imshow(img1_resize)
plt.title("Citra 1")
plt.axis("off")

# Citra 2
plt.subplot(2, 3, 2)
plt.imshow(img2_resize)
plt.title("Citra 2")
plt.axis("off")

# Hasil blending
for i, (a, b, desc, result) in enumerate(blend_results):
    plt.subplot(2, 3, i + 4)
    plt.imshow(result)
    plt.title(f"Blending α={a}, β={b}")
    plt.axis("off")

plt.suptitle("Image Blending - Perbandingan Hasil", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Tabel perbandingan hasil blending
print("=" * 70)
print(f"{"α":^10} | {"β":^10} | {"Analisis Hasil"}")
print("=" * 70)
print(f"{0.2:^10} | {0.8:^10} | Citra 2 (gelap) lebih dominan, hasil cenderung gelap")
print(f"{0.5:^10} | {0.5:^10} | Kedua citra seimbang, detail dari kedua citra terlihat")
print(f"{0.8:^10} | {0.2:^10} | Citra 1 (terang) lebih dominan, hasil cenderung terang")
print("=" * 70)
```

## Pertanyaan Analisis - Image Blending

### 1. Apa pengaruh nilai α dan β terhadap hasil blending?

Nilai α dan β menentukan proporsi kontribusi masing-masing citra terhadap hasil akhir.
α adalah bobot untuk citra pertama (I1), β adalah bobot untuk citra kedua (I2).
Semakin besar α, semakin dominan citra pertama dalam hasil blending, dan sebaliknya.
Karena α + β = 1, keduanya saling melengkapi.

### 2. Citra mana yang lebih dominan ketika α lebih besar?

Ketika α lebih besar dari β, maka **citra pertama (I1)** lebih dominan dalam hasil blending.
Sebaliknya, ketika β lebih besar, **citra kedua (I2)** yang lebih dominan.

### 3. Apa yang terjadi jika α + β > 1?

Jika α + β > 1, nilai pixel hasil blending bisa melebihi rentang maksimum (255).
Hal ini dapat menyebabkan **saturasi/clipping** di mana area terang menjadi putih total (255),
sehingga detail pada area tersebut hilang. Sebaliknya, jika α + β < 1, hasil blending akan
cenderung lebih gelap dari yang diharapkan.

### 4. Dalam aplikasi nyata, image blending dapat digunakan untuk apa?

Image blending memiliki banyak aplikasi nyata, antara lain:
- **Watermarking**: Menambahkan logo/teks transparan ke atas citra
- **Efek transisi**: Membuat transisi halus antara dua foto (slideshow)
- **Augmented reality**: Menggabungkan objek virtual dengan dunia nyata
- **Komposisi foto**: Menggabungkan beberapa eksposur (HDR imaging)
- **Desain grafis**: Membuat efek double exposure atau overlay tekstur

### 1. Apa yang terjadi pada pixel terang setelah dilakukan operasi negatif?

Pixel terang (nilai mendekati 255) akan berubah menjadi pixel gelap (nilai mendekati 0).
Misalnya pixel dengan nilai 200 akan menjadi 55 setelah operasi negatif (255 - 200 = 55).

### 2. Apa yang terjadi pada pixel gelap setelah dilakukan operasi negatif?

Pixel gelap (nilai mendekati 0) akan berubah menjadi pixel terang (nilai mendekati 255).
Misalnya pixel dengan nilai 30 akan menjadi 225 setelah operasi negatif (255 - 30 = 225).

Histogram akan **tercermin (mirror)** pada sumbu horizontal. Nilai-nilai pixel akan terbalik:
- Frekuensi di nilai 0 akan berpindah ke nilai 255
- Frekuensi di nilai 255 akan berpindah ke nilai 0
- Frekuensi di nilai x akan berpindah ke nilai (255 - x)
Bentuk histogram secara keseluruhan akan sama tetapi terbalik arah.

### 4. Dalam kondisi apa image negative dapat berguna?

Image negative berguna dalam beberapa kondisi:
- **Analisis citra medis**: Membantu melihat detail pada citra X-ray atau MRI
- **Deteksi tepi**: Memudahkan identifikasi tepi objek dalam citra
- **Peningkatan detail**: Area gelap yang tadinya sulit dilihat menjadi terang
- **Fotografi kreatif**: Efek artistik untuk menghasilkan tampilan vintage atau film negatif
- **Citra dengan dominasi gelap**: Membantu melihat detail yang tersembunyi di area gelap

## Bagian E: Transformasi Logaritmik

Transformasi logaritmik menggunakan persamaan: **s = c * log(1 + r)**

dengan:
- r = nilai pixel input
- s = nilai pixel output
- c = 255 / log(1 + r_max) = konstanta skala
- r_max = nilai pixel maksimum pada citra

Transformasi ini memperkuat area gelap dan mengkompresi area terang.

```python
# Transformasi logaritmik untuk citra 1
c1 = 255 / np.log10(1 + np.max(img1_resize))
img1_log = c1 * np.log10(img1_resize + 1)
img1_log = np.array(img1_log, dtype=np.uint8)

# Transformasi logaritmik untuk citra 2
c2 = 255 / np.log10(1 + np.max(img2_resize))
img2_log = c2 * np.log10(img2_resize + 1)
img2_log = np.array(img2_log, dtype=np.uint8)

print(f"Konstanta skala c untuk citra 1: {c1:.4f}")
print(f"Konstanta skala c untuk citra 2: {c2:.4f}")
```

```python
# Visualisasi citra asli vs transformasi logaritmik
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
plt.imshow(img1_resize)
plt.title("Citra 1 - Asli")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(img1_log)
plt.title("Citra 1 - Log Transform")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(img2_resize)
plt.title("Citra 2 - Asli")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(img2_log)
plt.title("Citra 2 - Log Transform")
plt.axis("off")

plt.suptitle("Perbandingan Citra Asli dan Transformasi Logaritmik", fontsize=14)
plt.tight_layout()
plt.show()
```

## Pertanyaan Analisis - Transformasi Logaritmik

### 1. Apa efek transformasi logaritmik terhadap area gelap pada citra?

Transformasi logaritmik **memperkuat area gelap**. Karena fungsi logaritmik memiliki kemiringan (slope)
yang lebih curam pada nilai input rendah, pixel-pixel gelap akan mengalami peningkatan brightness
yang signifikan. Detail-detail yang sebelumnya tersembunyi di area gelap menjadi lebih terlihat.

### 2. Apa efek transformasi logaritmik terhadap area terang pada citra?

Transformasi logaritmik **mengkompresi area terang**. Karena fungsi logaritmik memiliki kemiringan
yang lebih landai pada nilai input tinggi, perbedaan antara pixel-pixel terang menjadi lebih kecil.
Akibatnya, detail pada area terang bisa berkurang atau mengalami penurunan kontras.

### 3. Mengapa transformasi logaritmik sering digunakan untuk meningkatkan detail pada area gelap?

Karena sifat fungsi logaritmik yang responsif terhadap perubahan kecil pada nilai pixel rendah.
Rentang nilai gelap (0-50) akan diregangkan ke rentang yang lebih lebar, sehingga perbedaan kecil
antar pixel menjadi lebih terlihat oleh mata manusia. Ini sangat berguna untuk citra yang underexposed
atau memiliki area gelap dengan detail samar.

### 4. Apa risiko transformasi logaritmik jika digunakan pada citra yang sudah sangat terang?

Jika diterapkan pada citra yang sudah sangat terang, transformasi logaritmik justru dapat:
- **Mengurangi kontras** pada area terang karena kompresi nilai pixel
- **Membuat citra terlihat washed out** karena nilai-nilai pixel tinggi dipetakan ke rentang yang lebih sempit
- **Menghilangkan detail halus** pada area dengan pencahayaan tinggi
Transformasi logaritmik paling efektif untuk citra dengan rentang dinamis lebar atau area gelap yang dominan.

## Bagian F: Transformasi Gamma

Transformasi gamma menggunakan persamaan: **s = r^γ**

Langkah-langkah:
1. Normalisasi nilai pixel ke rentang [0, 1] dengan membagi 255.0
2. Terapkan pangkat gamma: s = r^γ
3. Kembalikan ke rentang [0, 255] dan konversi ke uint8

Kita akan mencoba 4 nilai gamma: **0.5, 1.0, 2.0, 2.5**

```python
# Fungsi transformasi gamma
def apply_gamma(img, gamma):
    img_norm = img / 255.0
    img_gamma = img_norm ** gamma
    img_gamma = np.uint8(img_gamma * 255)
    return img_gamma

# Nilai gamma yang akan diuji
gamma_values = [0.5, 1.0, 2.0, 2.5]

# Terapkan untuk citra 1 dan 2
img1_gamma_results = {}
img2_gamma_results = {}

for g in gamma_values:
    img1_gamma_results[g] = apply_gamma(img1_resize, g)
    img2_gamma_results[g] = apply_gamma(img2_resize, g)
    print(f"Gamma = {g}: selesai")
```

```python
# Visualisasi transformasi gamma untuk citra 1
plt.figure(figsize=(15, 8))

for i, g in enumerate(gamma_values):
    plt.subplot(1, 4, i + 1)
    plt.imshow(img1_gamma_results[g])
    plt.title(f"Gamma = {g}")
    plt.axis("off")

plt.suptitle("Citra 1 - Transformasi Gamma", fontsize=14)
plt.tight_layout()
plt.show()
```

## Pertanyaan Analisis - Transformasi Gamma

### 1. Apa yang terjadi ketika γ < 1?

Ketika γ < 1, fungsi gamma memetakan nilai pixel rendah ke nilai yang lebih tinggi secara proporsional.
Hasilnya: **citra menjadi lebih terang**, terutama pada area gelap. Kontras pada area gelap meningkat
karena rentang nilai gelap diregangkan. Detail pada bayangan menjadi lebih terlihat.

### 2. Apa yang terjadi ketika γ = 1?

Ketika γ = 1, transformasi gamma tidak mengubah citra sama sekali karena s = r^1 = r.
Ini berfungsi sebagai **referensi/kontrol** untuk membandingkan dengan nilai gamma lainnya.

### 3. Apa yang terjadi ketika γ > 1?

Ketika γ > 1, efeknya kebalikan dari γ < 1. Fungsi gamma memetakan nilai pixel ke nilai yang lebih rendah.
Hasilnya: **citra menjadi lebih gelap**. Area terang dikompresi dan area gelap semakin gelap.
Semakin besar γ, semakin gelap hasilnya.

### 4. Bagaimana transformasi gamma dapat digunakan untuk koreksi pencahayaan?

Transformasi gamma dapat digunakan untuk:
- **Citra terlalu gelap (underexposed)**: Gunakan γ < 1 untuk mencerahkan area gelap
- **Citra terlalu terang (overexposed)**: Gunakan γ > 1 untuk mengurangi kecerahan
- **Koreksi gamma monitor**: Menyesuaikan citra agar tampak natural di berbagai perangkat
Gamma correction adalah standar dalam fotografi digital dan rendering untuk mengkompensasi
karakteristik non-linear dari layar dan mata manusia.

### 5. Kapan transformasi gamma lebih cocok digunakan dibanding transformasi logaritmik?

Transformasi gamma lebih cocok ketika:
- Kita ingin **kontrol yang lebih fleksibel** melalui pemilihan nilai γ
- Kita perlu **menyesuaikan tingkat koreksi** secara bertahap (nilai γ kontinu)
- Kita ingin **mencerahkan atau menggelapkan** citra secara keseluruhan (bukan hanya area gelap)
- Kita bekerja dengan **citra yang sudah terkalibrasi** (seperti sRGB) di mana gamma sudah distandarisasi

Sementara transformasi logaritmik lebih cocok untuk **meningkatkan detail pada area gelap** secara spesifik
tanpa banyak memengaruhi area terang.

Teknik yang paling tepat adalah **Transformasi Gamma dengan γ < 1** (misalnya γ = 0.5).

logaritmik juga bisa digunakan jika fokus utamanya adalah detail di area gelap.

**2. Jika citra terlalu terang, teknik mana yang paling tepat digunakan? Jelaskan.**

Teknik yang paling tepat adalah **Transformasi Gamma dengan γ > 1** (misalnya γ = 2.0 atau 2.5).
Alasan: γ > 1 akan menggelapkan citra, mengurangi kecerahan berlebih, dan membantu mengembalikan
detail pada area yang tadinya terlalu terang (overexposed). Transformasi logaritmik kurang tepat
karena justru akan mengkompresi area terang lebih lanjut.

**3. Jika ingin menggabungkan dua citra, teknik mana yang digunakan? Jelaskan.**

Teknik yang digunakan adalah **Image Blending** dengan fungsi `cv2.addWeighted()`.
Alasan: Image blending menggabungkan dua citra secara linier dengan bobot α dan β. Dengan mengatur
bobot, kita bisa mengontrol seberapa dominan masing-masing citra dalam hasil akhir. Teknik lain
tidak dirancang untuk menggabungkan dua citra.

**4. Jika ingin menonjolkan detail pada area gelap, apakah lebih tepat menggunakan log transform atau gamma transform? Jelaskan.**

**Transformasi Logaritmik** lebih tepat untuk menonjolkan detail pada area gelap.
Alasan: Fungsi logaritmik memiliki respons yang lebih agresif terhadap nilai pixel rendah dibanding
gamma. Slope kurva logaritmik sangat curam di area gelap, sehingga perbedaan kecil antar pixel gelap
menjadi sangat terlihat. Transformasi gamma dengan γ < 1 juga mencerahkan area gelap, tetapi efeknya
lebih merata di seluruh rentang dan kurang agresif dalam meregangkan detail gelap.

**5. Apakah semua transformasi citra selalu memperbaiki kualitas citra? Jelaskan.**

**Tidak.** Tidak semua transformasi selalu memperbaiki kualitas citra. Setiap transformasi memiliki
trade-off:
- Image blending bisa mengurangi kontras jika bobot tidak tepat
- Image negative tidak memperbaiki kualitas, hanya membalikkan nilai pixel
- Transformasi logaritmik dapat mengurangi detail pada area terang
- Transformasi gamma dengan nilai ekstrem dapat menyebabkan hilangnya detail (clipping)
Pemilihan teknik harus disesuaikan dengan masalah spesifik dan karakteristik citra.

## Bagian H: Refleksi Pribadi

**1. Teknik manipulasi citra mana yang paling mudah dipahami? Mengapa?**

**Image negative** adalah teknik yang paling mudah dipahami karena operasinya sangat sederhana,
yaitu mengurangi nilai pixel dari 255 (I_negatif = 255 - I). Konsepnya intuitif: terang menjadi gelap
dan sebaliknya, seperti film negatif. Tidak perlu perhitungan rumit atau konstanta tambahan.

**2. Teknik mana yang paling sulit dianalisis? Mengapa?**

**Image blending** cukup sulit dianalisis karena melibatkan kombinasi linear dari dua citra dengan
bobot yang berbeda. Menentukan bobot optimal untuk menghasilkan efek yang diinginkan membutuhkan
eksperimen dan pemahaman tentang bagaimana kontribusi masing-masing citra memengaruhi hasil akhir.
Selain itu, interpretasi hasil blending bersifat subjektif tergantung tujuan yang ingin dicapai.

**3. Apa perbedaan efek transformasi logaritmik dan transformasi gamma?**

Transformasi logaritmik menggunakan fungsi logaritma (s = c * log(1 + r)) yang bersifat spesifik:
sangat responsif terhadap perubahan di area gelap tetapi kurang responsif di area terang.
Transformasi gamma menggunakan fungsi pangkat (s = r^γ) yang lebih fleksibel: dengan mengubah nilai γ,
kita bisa mencerahkan (γ < 1) atau menggelapkan (γ > 1) citra secara keseluruhan.
Logaritmik lebih cocok untuk detail gelap, gamma lebih cocok untuk koreksi pencahayaan umum.

memiliki trade-off: meningkatkan satu aspek (misal detail gelap) bisa mengorbankan aspek lain
