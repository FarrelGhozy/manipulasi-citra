## Bagian A: Membaca dan Menampilkan Citra

Pada bagian ini kita akan:
1. Membaca dua citra berbeda menggunakan OpenCV
2. Melakukan konversi warna BGR ke RGB
3. Menampilkan informasi dasar citra (shape, dtype, min/max pixel)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print(f"OpenCV version: {cv2.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# Membaca dua citra
img1 = cv2.imread("images/image1.jpg")
img2 = cv2.imread("images/image2.jpg")

if img1 is None:
    print("ERROR: image1.jpg tidak terbaca!")
else:
    print(f"image1.jpg berhasil dibaca: shape={img1.shape}")

if img2 is None:
    print("ERROR: image2.jpg tidak terbaca!")
else:
    print(f"image2.jpg berhasil dibaca: shape={img2.shape}")
```

```python
# Menampilkan citra sebelum konversi (masih BGR)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img1)  # Akan terlihat aneh karena BGR
plt.title("Citra 1 - BGR (belum dikonversi)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2)  # Akan terlihat aneh karena BGR
plt.title("Citra 2 - BGR (belum dikonversi)")
plt.axis("off")

plt.suptitle("Citra Sebelum Konversi BGR ke RGB (warna tidak sesuai)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Konversi BGR ke RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

print("Konversi BGR ke RGB berhasil!")
```

```python
# Menampilkan citra setelah konversi ke RGB
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img1_rgb)
plt.title("Citra 1 - RGB (setelah konversi)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2_rgb)
plt.title("Citra 2 - RGB (setelah konversi)")
plt.axis("off")

plt.suptitle("Citra Setelah Konversi BGR ke RGB (warna sesuai)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Informasi detail citra
for name, img in [("image1.jpg", img1), ("image2.jpg", img2)]:
    print(f"=== {name} ===")
    print(f"  Shape (H, W, C): {img.shape}")
    print(f"  Tinggi (pixels): {img.shape[0]}")
    print(f"  Lebar (pixels): {img.shape[1]}")
    print(f"  Channel: {img.shape[2]}")
    print(f"  Tipe data: {img.dtype}")
    print(f"  Nilai min pixel: {img.min()}")
    print(f"  Nilai max pixel: {img.max()}")
    print()
```

## Pertanyaan Analisis

### 1. Mengapa warna citra dapat terlihat tidak sesuai jika langsung ditampilkan setelah dibaca dengan OpenCV?

OpenCV membaca citra dalam format **BGR** (Blue-Green-Red), sedangkan Matplotlib menggunakan format **RGB** (Red-Green-Blue).
Ketika citra BGR ditampilkan dengan `plt.imshow()`, Matplotlib menginterpretasikan channel pertama sebagai Red (padahal itu Blue),
channel kedua sebagai Green (tetap Green), dan channel ketiga sebagai Blue (padahal itu Red). Akibatnya, warna biru dan merah tertukar
sehingga citra terlihat tidak natural.

### 2. Apa perbedaan BGR dan RGB?

BGR dan RGB adalah dua format urutan channel warna yang berbeda:
- **RGB**: Urutan channel Red → Green → Blue (umum digunakan di monitor, web, Matplotlib)
- **BGR**: Urutan channel Blue → Green → Red (digunakan oleh OpenCV secara default)

Keduanya menyimpan informasi warna yang sama, hanya urutan penyimpanannya yang berbeda.

### 3. Mengapa penting memahami format warna sebelum melakukan manipulasi citra?

Penting karena jika kita melakukan manipulasi (seperti mengakses channel tertentu, mengubah warna, atau menampilkan citra)
tanpa memperhatikan format warnanya, hasilnya bisa salah atau tidak sesuai yang diharapkan. Misalnya, jika kita ingin
memproses channel merah tetapi data tersimpan di channel ketiga (BGR), maka yang diproses adalah channel biru.
Memahami format warna memastikan setiap operasi berjalan pada channel yang benar.

## Bagian B: Resize Citra

Pada bagian ini kita akan melakukan resize dua citra agar memiliki ukuran yang sama.
Ini diperlukan sebelum melakukan image blending di bagian selanjutnya.

```python
# Menampilkan ukuran sebelum resize
print("Ukuran sebelum resize:")
print(f"  image1.jpg: {img1_rgb.shape[0]} x {img1_rgb.shape[1]} pixels")
print(f"  image2.jpg: {img2_rgb.shape[0]} x {img2_rgb.shape[1]} pixels")
print()

# Menentukan ukuran target (contoh: 600x400)
target_w, target_h = 600, 400
print(f"Ukuran target resize: {target_w} x {target_h} pixels")
```

```python
# Melakukan resize
img1_resize = cv2.resize(img1_rgb, (target_w, target_h))
img2_resize = cv2.resize(img2_rgb, (target_w, target_h))

print("Setelah resize:")
print(f"  image1.jpg: {img1_resize.shape[0]} x {img1_resize.shape[1]} pixels")
print(f"  image2.jpg: {img2_resize.shape[0]} x {img2_resize.shape[1]} pixels")
```

```python
# Visualisasi sebelum dan sesudah resize
plt.figure(figsize=(14, 10))

# Citra 1 sebelum resize
plt.subplot(2, 2, 1)
plt.imshow(img1_rgb)
plt.title(f"Citra 1 - Sebelum resize ({img1_rgb.shape[1]}x{img1_rgb.shape[0]})")
plt.axis("off")

# Citra 1 setelah resize
plt.subplot(2, 2, 2)
plt.imshow(img1_resize)
plt.title(f"Citra 1 - Setelah resize ({target_w}x{target_h})")
plt.axis("off")

# Citra 2 sebelum resize
plt.subplot(2, 2, 3)
plt.imshow(img2_rgb)
plt.title(f"Citra 2 - Sebelum resize ({img2_rgb.shape[1]}x{img2_rgb.shape[0]})")
plt.axis("off")

# Citra 2 setelah resize
plt.subplot(2, 2, 4)
plt.imshow(img2_resize)
plt.title(f"Citra 2 - Setelah resize ({target_w}x{target_h})")
plt.axis("off")

plt.suptitle("Perbandingan Sebelum dan Sesudah Resize", fontsize=14)
plt.tight_layout()
plt.show()
```

## Pertanyaan Analisis - Resize

### 1. Mengapa dua citra harus memiliki ukuran yang sama sebelum dilakukan image blending?

Image blending adalah operasi aritmatika per-pixel. Operasi `I_blend = α*I1 + β*I2 + γ` membutuhkan
setiap pixel di posisi (x, y) pada citra pertama dijumlahkan dengan pixel di posisi (x, y) yang sama pada
citra kedua. Jika ukuran kedua citra berbeda, maka operasi penjumlahan per-pixel tidak dapat dilakukan
karena dimensinya tidak cocok. OpenCV akan menolak operasi blending jika ukuran tidak sama.

### 2. Apa risiko resize terhadap kualitas citra?

Risiko utama resize terhadap kualitas citra adalah:
- **Jika diperbesar (upscaling)**: Citra bisa menjadi buram/pixelated karena algoritma interpolasi harus
  "menebak" nilai pixel baru yang tidak ada di citra asli. Detail halus bisa hilang.
- **Jika diperkecil (downscaling)**: Informasi detail bisa hilang karena pixel-pixel dijumlahkan/dirata-rata.
  Tekstur halus dan objek kecil mungkin menjadi tidak jelas atau hilang sama sekali.

### 3. Apakah resize dapat mengubah proporsi objek dalam citra? Jelaskan.

Ya, jika rasio aspek (aspect ratio) tidak dipertahankan saat resize. Misalnya citra asli 800x600 (rasio 4:3)
di-resize menjadi 600x400 (rasio 3:2). Dalam contoh ini, proporsi objek akan berubah karena rasio aspeknya
tidak sama. Objek akan terlihat lebih lebar atau lebih sempit dari aslinya. Untuk mempertahankan proporsi,
kita harus menjaga rasio aspek yang sama atau melakukan padding/cropping.

Pada eksperimen ini, citra 1 (600x386) di-resize ke (600x400), sehingga proporsinya sedikit berubah.

## Bagian C: Image Blending

Image blending adalah teknik menggabungkan dua citra dengan bobot tertentu.
Persamaan: **I_blend = α * I1 + β * I2 + γ** dengan **α + β = 1**

Kita akan mencoba 3 kombinasi bobot:

```python
# Menentukan kombinasi bobot
weights = [
    (0.2, 0.8, "α=0.2, β=0.8 (Citra 2 lebih dominan)"),
    (0.5, 0.5, "α=0.5, β=0.5 (Seimbang)"),
    (0.8, 0.2, "α=0.8, β=0.2 (Citra 1 lebih dominan)")
]

# Melakukan blending untuk setiap kombinasi
blend_results = []
for a, b, desc in weights:
    result = cv2.addWeighted(img1_resize, a, img2_resize, b, 0)
    blend_results.append((a, b, desc, result))
    print(f"Blending: {desc}")

print("\nSemua blending berhasil!")
```

```python
# Visualisasi citra asli dan hasil blending
plt.figure(figsize=(15, 10))

# Citra 1
plt.subplot(2, 3, 1)
plt.imshow(img1_resize)
plt.title("Citra 1")
plt.axis("off")

# Citra 2
plt.subplot(2, 3, 2)
plt.imshow(img2_resize)
plt.title("Citra 2")
plt.axis("off")

# Hasil blending
for i, (a, b, desc, result) in enumerate(blend_results):
    plt.subplot(2, 3, i + 4)
    plt.imshow(result)
    plt.title(f"Blending α={a}, β={b}")
    plt.axis("off")

plt.suptitle("Image Blending - Perbandingan Hasil", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Tabel perbandingan hasil blending
print("=" * 70)
print(f"{"α":^10} | {"β":^10} | {"Analisis Hasil"}")
print("=" * 70)
print(f"{0.2:^10} | {0.8:^10} | Citra 2 (gelap) lebih dominan, hasil cenderung gelap")
print(f"{0.5:^10} | {0.5:^10} | Kedua citra seimbang, detail dari kedua citra terlihat")
print(f"{0.8:^10} | {0.2:^10} | Citra 1 (terang) lebih dominan, hasil cenderung terang")
print("=" * 70)
```

## Pertanyaan Analisis - Image Blending

### 1. Apa pengaruh nilai α dan β terhadap hasil blending?

Nilai α dan β menentukan proporsi kontribusi masing-masing citra terhadap hasil akhir.
α adalah bobot untuk citra pertama (I1), β adalah bobot untuk citra kedua (I2).
Semakin besar α, semakin dominan citra pertama dalam hasil blending, dan sebaliknya.
Karena α + β = 1, keduanya saling melengkapi.

### 2. Citra mana yang lebih dominan ketika α lebih besar?

Ketika α lebih besar dari β, maka **citra pertama (I1)** lebih dominan dalam hasil blending.
Sebaliknya, ketika β lebih besar, **citra kedua (I2)** yang lebih dominan.

### 3. Apa yang terjadi jika α + β > 1?

Jika α + β > 1, nilai pixel hasil blending bisa melebihi rentang maksimum (255).
Hal ini dapat menyebabkan **saturasi/clipping** di mana area terang menjadi putih total (255),
sehingga detail pada area tersebut hilang. Sebaliknya, jika α + β < 1, hasil blending akan
cenderung lebih gelap dari yang diharapkan.

### 4. Dalam aplikasi nyata, image blending dapat digunakan untuk apa?

Image blending memiliki banyak aplikasi nyata, antara lain:
- **Watermarking**: Menambahkan logo/teks transparan ke atas citra
- **Efek transisi**: Membuat transisi halus antara dua foto (slideshow)
- **Augmented reality**: Menggabungkan objek virtual dengan dunia nyata
- **Komposisi foto**: Menggabungkan beberapa eksposur (HDR imaging)
- **Desain grafis**: Membuat efek double exposure atau overlay tekstur

## Bagian D: Image Negative

Image negative adalah teknik membalikkan nilai pixel dengan persamaan:
**I_negatif = 255 - I**

Teknik ini menghasilkan efek "film negatif" di mana area terang menjadi gelap dan sebaliknya.

```python
# Membuat citra negatif untuk citra 1
img1_negative = 255 - img1_resize

# Membuat citra negatif untuk citra 2
img2_negative = 255 - img2_resize

print("Citra negatif berhasil dibuat!")
```

```python
# Visualisasi citra asli vs negatif
plt.figure(figsize=(14, 8))

# Citra 1 asli
plt.subplot(2, 2, 1)
plt.imshow(img1_resize)
plt.title("Citra 1 - Asli")
plt.axis("off")

# Citra 1 negatif
plt.subplot(2, 2, 2)
plt.imshow(img1_negative)
plt.title("Citra 1 - Negatif")
plt.axis("off")

# Citra 2 asli
plt.subplot(2, 2, 3)
plt.imshow(img2_resize)
plt.title("Citra 2 - Asli")
plt.axis("off")

# Citra 2 negatif
plt.subplot(2, 2, 4)
plt.imshow(img2_negative)
plt.title("Citra 2 - Negatif")
plt.axis("off")

plt.suptitle("Perbandingan Citra Asli dan Citra Negatif", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Plot histogram untuk citra 1
plt.figure(figsize=(14, 6))

# Histogram citra 1 asli
plt.subplot(1, 2, 1)
colors = ("red", "green", "blue")
for i, color in enumerate(colors):
    hist = cv2.calcHist([img1_resize], [i], None, [256], [0, 256])
    plt.plot(hist, color=color, alpha=0.7)
plt.title("Histogram Citra 1 - Asli")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.xlim([0, 255])

# Histogram citra 1 negatif
plt.subplot(1, 2, 2)
for i, color in enumerate(colors):
    hist = cv2.calcHist([img1_negative], [i], None, [256], [0, 256])
    plt.plot(hist, color=color, alpha=0.7)
plt.title("Histogram Citra 1 - Negatif")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.xlim([0, 255])

plt.tight_layout()
plt.show()
```

```python
# Plot histogram untuk citra 2
plt.figure(figsize=(14, 6))

# Histogram citra 2 asli
plt.subplot(1, 2, 1)
for i, color in enumerate(colors):
    hist = cv2.calcHist([img2_resize], [i], None, [256], [0, 256])
    plt.plot(hist, color=color, alpha=0.7)
plt.title("Histogram Citra 2 - Asli")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.xlim([0, 255])

# Histogram citra 2 negatif
plt.subplot(1, 2, 2)
for i, color in enumerate(colors):
    hist = cv2.calcHist([img2_negative], [i], None, [256], [0, 256])
    plt.plot(hist, color=color, alpha=0.7)
plt.title("Histogram Citra 2 - Negatif")
plt.xlabel("Nilai Pixel")
plt.ylabel("Frekuensi")
plt.xlim([0, 255])

plt.tight_layout()
plt.show()
```

## Pertanyaan Analisis - Image Negative

### 1. Apa yang terjadi pada pixel terang setelah dilakukan operasi negatif?

Pixel terang (nilai mendekati 255) akan berubah menjadi pixel gelap (nilai mendekati 0).
Misalnya pixel dengan nilai 200 akan menjadi 55 setelah operasi negatif (255 - 200 = 55).

### 2. Apa yang terjadi pada pixel gelap setelah dilakukan operasi negatif?

Pixel gelap (nilai mendekati 0) akan berubah menjadi pixel terang (nilai mendekati 255).
Misalnya pixel dengan nilai 30 akan menjadi 225 setelah operasi negatif (255 - 30 = 225).

### 3. Bagaimana perubahan histogram setelah citra dibuat negatif?

Histogram akan **tercermin (mirror)** pada sumbu horizontal. Nilai-nilai pixel akan terbalik:
- Frekuensi di nilai 0 akan berpindah ke nilai 255
- Frekuensi di nilai 255 akan berpindah ke nilai 0
- Frekuensi di nilai x akan berpindah ke nilai (255 - x)
Bentuk histogram secara keseluruhan akan sama tetapi terbalik arah.

### 4. Dalam kondisi apa image negative dapat berguna?

Image negative berguna dalam beberapa kondisi:
- **Analisis citra medis**: Membantu melihat detail pada citra X-ray atau MRI
- **Deteksi tepi**: Memudahkan identifikasi tepi objek dalam citra
- **Peningkatan detail**: Area gelap yang tadinya sulit dilihat menjadi terang
- **Fotografi kreatif**: Efek artistik untuk menghasilkan tampilan vintage atau film negatif
- **Citra dengan dominasi gelap**: Membantu melihat detail yang tersembunyi di area gelap
