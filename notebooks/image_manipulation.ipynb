## Bagian A: Membaca dan Menampilkan Citra

Pada bagian ini kita akan:
1. Membaca dua citra berbeda menggunakan OpenCV
2. Melakukan konversi warna BGR ke RGB
3. Menampilkan informasi dasar citra (shape, dtype, min/max pixel)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

print(f"OpenCV version: {cv2.__version__}")
print(f"NumPy version: {np.__version__}")
```

```python
# Membaca dua citra
img1 = cv2.imread("images/image1.jpg")
img2 = cv2.imread("images/image2.jpg")

if img1 is None:
    print("ERROR: image1.jpg tidak terbaca!")
else:
    print(f"image1.jpg berhasil dibaca: shape={img1.shape}")

if img2 is None:
    print("ERROR: image2.jpg tidak terbaca!")
else:
    print(f"image2.jpg berhasil dibaca: shape={img2.shape}")
```

```python
# Menampilkan citra sebelum konversi (masih BGR)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img1)  # Akan terlihat aneh karena BGR
plt.title("Citra 1 - BGR (belum dikonversi)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2)  # Akan terlihat aneh karena BGR
plt.title("Citra 2 - BGR (belum dikonversi)")
plt.axis("off")

plt.suptitle("Citra Sebelum Konversi BGR ke RGB (warna tidak sesuai)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Konversi BGR ke RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

print("Konversi BGR ke RGB berhasil!")
```

```python
# Menampilkan citra setelah konversi ke RGB
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(img1_rgb)
plt.title("Citra 1 - RGB (setelah konversi)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img2_rgb)
plt.title("Citra 2 - RGB (setelah konversi)")
plt.axis("off")

plt.suptitle("Citra Setelah Konversi BGR ke RGB (warna sesuai)", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Informasi detail citra
for name, img in [("image1.jpg", img1), ("image2.jpg", img2)]:
    print(f"=== {name} ===")
    print(f"  Shape (H, W, C): {img.shape}")
    print(f"  Tinggi (pixels): {img.shape[0]}")
    print(f"  Lebar (pixels): {img.shape[1]}")
    print(f"  Channel: {img.shape[2]}")
    print(f"  Tipe data: {img.dtype}")
    print(f"  Nilai min pixel: {img.min()}")
    print(f"  Nilai max pixel: {img.max()}")
    print()
```

## Pertanyaan Analisis

### 1. Mengapa warna citra dapat terlihat tidak sesuai jika langsung ditampilkan setelah dibaca dengan OpenCV?

OpenCV membaca citra dalam format **BGR** (Blue-Green-Red), sedangkan Matplotlib menggunakan format **RGB** (Red-Green-Blue).
Ketika citra BGR ditampilkan dengan `plt.imshow()`, Matplotlib menginterpretasikan channel pertama sebagai Red (padahal itu Blue),
channel kedua sebagai Green (tetap Green), dan channel ketiga sebagai Blue (padahal itu Red). Akibatnya, warna biru dan merah tertukar
sehingga citra terlihat tidak natural.

### 2. Apa perbedaan BGR dan RGB?

BGR dan RGB adalah dua format urutan channel warna yang berbeda:
- **RGB**: Urutan channel Red → Green → Blue (umum digunakan di monitor, web, Matplotlib)
- **BGR**: Urutan channel Blue → Green → Red (digunakan oleh OpenCV secara default)

Keduanya menyimpan informasi warna yang sama, hanya urutan penyimpanannya yang berbeda.

### 3. Mengapa penting memahami format warna sebelum melakukan manipulasi citra?

Penting karena jika kita melakukan manipulasi (seperti mengakses channel tertentu, mengubah warna, atau menampilkan citra)
tanpa memperhatikan format warnanya, hasilnya bisa salah atau tidak sesuai yang diharapkan. Misalnya, jika kita ingin
memproses channel merah tetapi data tersimpan di channel ketiga (BGR), maka yang diproses adalah channel biru.
Memahami format warna memastikan setiap operasi berjalan pada channel yang benar.
